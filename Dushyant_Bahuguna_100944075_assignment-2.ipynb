```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_size = 100
```

```python
X1 = np.random.normal(5, 2, data_size).reshape(-1,1)
X2 = np.random.normal(10, 2, data_size).reshape(-1,1)

X1,X2
```

```python
w1_fixed = 4.8
w2_fixed = 2.5
w3_fixed = 1.2
```

```python
Y = w3_fixed*(w1_fixed*X1 + w2_fixed*X2)
Y
```

```python
#defining the function ANN

def my_ANN_model(x1, x2, y, Learning_Rate=0.001, nr_epochs=20, w1=-0.22, w2=0.17, w3=0.97, costs=[]):
    #Defining the list for cost
    for epoch in range(nr_epochs):
        H1 = w1 * x1 + w2 * x2 # Defining intermediate state H1
        Output = w3 * H1 # Defining output
        cost = np.mean((y-Output)**2) #MSE
        costs.append(cost) #Appending costs in the list
        
        # Finding the change in w
        delta_w1 = np.mean((y-Output)*(-w3 * x1))
        delta_w2 = np.mean((y-Output)*(-w3 * x2))
        delta_w3 = np.mean((y-Output)*(-H1))
        
        #Updating the weights
        w1 = w1 - Learning_Rate * delta_w1
        w2 = w2 - Learning_Rate * delta_w2
        w3 = w3 - Learning_Rate * delta_w3
    return w1, w2, w3, costs # weights associated with the two input parameters and a list containing the cost values for each training epoch
```

```python
w1, w2, w3, costs = my_ANN_model(X1, X2, Y)
```

```python
plt.plot(costs)
plt.xlabel("Epoch")
plt.ylabel("Cost (MSE)")
plt.title("Training Loss Over Time")
plt.show()
```
